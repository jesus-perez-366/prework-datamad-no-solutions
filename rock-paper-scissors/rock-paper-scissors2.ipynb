```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

gestures=['rock','paper','scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

print("Introdusca el numero de rondas que desea jugar, recuerde que debe ser un numero impar")
n_rounds=2 
while n_rounds % 2 == 0:
    n_rounds=int(input()) 
    if n_rounds % 2 == 0: 
        print("el numero que ha introducido no es un numero impar, por favor vuelva a introducir otro")

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

rounds_to_win=int((n_rounds/2)+1)

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

m=random.choice(gestures)

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

print("acontinuacion aparecera una opcion a escoger, si quieres esa coloca OK, si deseas otra opcion coloca NO")
t="NO"
while t == "NO":
    s=random.choice(gestures)
    print(s)
    t=input("coloque OK o NO  ",)
    t=t.upper()
print("es momento de saber quien fue el ganador")


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

if m == "rock" and s=="scissors":
    w=1
elif m == "rock" and s=="paper":
    w=2
elif m == "rock" and s=="rock":
    w=0
elif m == "scissors" and s=="scissors":
    w=0
elif m == "scissors" and s=="paper":
    w=1
elif m == "scissors" and s=="rocks":
    w=2
elif m == "paper" and s=="scissors":
    w=2
elif m == "paper" and s=="paper":
    w=0
elif m == "paper" and s=="rock":
    w=1
 
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
# Create two variables that accumulate the wins of each participant
cpu_score=0
player_score=0
TIE=0
print("el cpu saco ",m)
print("el jugador saco ",s)
if w == 1:
    cpu_score+=1
    print("el ganador de esta partida es el cpu")
elif w==2:
    player_score+=1
    print("el ganador de esta partida es el jugador")
else:
    TIE+=1
    print("hubo empate en esta partida")




# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

contador=0
while rounds_to_win !=cpu_score and rounds_to_win !=player_score and rounds_to_win !=TIE and n_rounds != contador:
    print("acontinuacion aparecera una opcion a escoger, si quieres esa coloca OK, si deseas otra opcion coloca NO")
    t="NO"
    while t == "NO":
        m=random.choice(gestures)
        s=random.choice(gestures)
        print(s)
        t=input("coloque OK o NO  ",)
        t=t.upper()
    print("es momento de saber quien fue el ganador")
    if m == "rock" and s=="scissors":
        w=1
    elif m == "rock" and s=="paper":
        w=2
    elif m == "rock" and s=="rock":
        w=0
    elif m == "scissors" and s=="scissors":
        w=0
    elif m == "scissors" and s=="paper":
        w=1
    elif m == "scissors" and s=="rocks":
        w=2
    elif m == "paper" and s=="scissors":
        w=2
    elif m == "paper" and s=="paper":
        w=0
    elif m == "paper" and s=="rock":
        w=1
    print("el cpu saco ",m)
    print("el jugador saco ",s)
    if w == 1:
        cpu_score+=1
        print("el ganador de esta partida es el cpu")
    elif w==2:
        player_score+=1
        print("el ganador de esta partida es el jugador")
    else:
        TIE+=1
        print("hubo empate en esta partida")
    contador+=1

    
# Print by console the winner of the game based on who has more accumulated wins

if rounds_to_win ==cpu_score:
    print("el ganador final del juego es el cpu")
elif rounds_to_win ==player_score:
    print("el ganador final del juego es el jugador")
elif cpu_score>player_score:
    print("hubo ",TIE," empates, pero en el conteo final, el cpu gano mas partidas que el jugador, por ello el ganador es el cpu")
elif cpu_score<player_score:
    print("hubo ",TIE," empates, pero en el conteo final el jugador gano mas partidas que el cpu, por ello el ganador es el jugado")
else:
    print("se jugaron todas las rondas y no se puede determinar un ganador, por lo tanto este juego quedo en empate")
  
```
